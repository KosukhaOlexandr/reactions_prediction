```python
# these are standard telegram reactions emojis
# clown emoji was premium reaction until Sep 18, but still it was very popular

recorded_emoji = ['ü§°','üëç', 'üëé', '‚ù§', 'üî•', 'ü•∞', 'üëè', 'üòÅ', 'ü§î', 'ü§Ø', 'üò±', 'ü§¨', 'üò¢', 'üéâ', 'ü§©', 'ü§Æ', 'üí©']
recorded_emoji_set = set(recorded_emoji)
```

```python
empty_reactions_indexes = [] # don't see any other way but to use global variable for this, walrus(:=) is not working in colab

def standardize_reaction(data):
  i = 0
  #empty_reactions_indexes = []
  for index, val in data.reactions.items():
    i += 1
    #print(i)
    s = str(val).replace('None', 'null')
    s = s.replace('False', 'false')
    s = s.replace('True', 'true')
    s = s.replace('\'', '\"')
    s = s.replace('\\', '')
    reaction = pd.read_json(s, orient = 'records', encoding_errors = 'backslashreplace')

    if reaction.empty:
      empty_reactions_indexes.append(i - 1)
      standardized_reaction = pd.DataFrame(data=0, 
                                        index = ['count'], columns=recorded_emoji)
    else:
      reaction.drop(columns = ['_client', 'chosen'], inplace = True)
      reaction = reaction.T
      reaction.columns = reaction.iloc[0]
      reaction.drop('emoji', inplace = True)

      standardized_reaction = pd.DataFrame(data=0, 
                                          index = ['count'], columns=recorded_emoji)
      for emj, cnt in reaction.iteritems():
        if emj in recorded_emoji_set:
          standardized_reaction[emj] = cnt
      standardized_reaction.loc['count'] /= standardized_reaction.loc['count'].sum()
    
    yield standardized_reaction
    #
```

```python
reactions_divided = pd.DataFrame()

reactions_divided = pd.concat([std_reaction for std_reaction in standardize_reaction(data)], copy=False)
```

```python
def preprocess_text(text):
  import re
  import string

  new_text = emoji.replace_emoji(text, replace = '')
  new_text = new_text.replace('\n', ' ')
  new_text = new_text.lower()
  new_text = re.sub(r'http\S+', '', new_text)

  # deleting all punctuation
  mappingtbl = new_text.maketrans({punct: ' ' for punct in string.punctuation + '‚Äì‚Äî¬´¬ª‚Ññ‚Ç§‚Ç¨'})
  new_text = new_text.translate(mappingtbl)
  
  # removing multiple spaces
  new_text = ' '.join(new_text.split())
  
  # There can be text with hyperlinks in the end of news
  # that have such content: "–ø—ñ–¥–ø–∏—Å–∞—Ç–∏—Å—è youtube —Ç—Å–Ω –ø—ñ–¥—Ç—Ä–∏–º–∞—Ç–∏ —Ç—Å–Ω ua"
  # we need to delete it as well
  new_text = new_text.replace('–ø—ñ–¥–ø–∏—Å–∞—Ç–∏—Å—è youtube —Ç—Å–Ω –ø—ñ–¥—Ç—Ä–∏–º–∞—Ç–∏ —Ç—Å–Ω ua', ' ')

  # removing multiple spaces second time
  new_text = ' '.join(new_text.split())

  return new_text
```
